```python
# load necessary libraries
from qiskit import *
from qiskit import visualization
%matplotlib inline
import math
```

# 7.4.1 使用 ID 閘的量子電路 (ID gate)

```python
q=QuantumRegister(1)
circuit=QuantumCircuit(q)
circuit.id(q[0])
circuit.draw()
simulator=Aer.get_backend('statevector_simulator')
job=execute(circuit, simulator)
result=job.result()
statevector=result.get_statevector()
print(statevector)
```

```python
visualization.plot_bloch_multivector(statevector)
```
